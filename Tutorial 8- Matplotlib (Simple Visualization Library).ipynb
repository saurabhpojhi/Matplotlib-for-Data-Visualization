# Bar plot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color ='g')
plt.title('Bar graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

# Histograms

```python
a = np.array([22,87,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

 # Box plot using Matplot

```python
data =[np.random.normal(0, std, 100) for std in range(1,4)]

# rectengal box plot

plt.boxplot(data,vert=True,patch_artist=True)
```

 # Pie Chart

```python
# data to plot

labels = 'python','java','ruby','c++'
sizes = [215,130,245,210]
colors=['gold','yellow','lightcoral','lightskyblue']
explode = (0.1,0,0,0)  #explode 1st slice

# plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct= '%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
