```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
 # I have created two python list x and y . 
    
x = [1,2,3,4,5,6,7]          # x number of the weekend and y is number of temprature
y = [50,51,52,48,47,49,46]
    
```

```python
plt.xlabel('day')
plt.ylabel('temperature')
plt.title('wheather')
plt.plot(x,y, color='green', linewidth=5, linestyle='dotted')
```

 # Bar Chart

```python
Company = ['Google','Amazon','FB','Microsoft']
Revenue = [90,136,88,66]
Profit = [40,2,34,22]
```

```python
ypos = np.arange(len(Company))
ypos
```

```python
plt.xticks(ypos,Company)
plt.xlabel('revenue Bin')
plt.title('US Tech stocks')
plt.bar(ypos,Revenue, label='Revenue')
plt.bar(ypos,Profit, label='Profit')
plt.legend()
```

 # Histogram

```python
blood_sugar_men = [112,65,34,65,48,98,58,100,132,116,47,98]
blood_sugar_women = [111,67,39,65,72,88,98,135,122,136,89,68]

plt.xlabel('range')
plt.ylabel('total no of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men,blood_sugar_women], bins = [80,100,125,150], rwidth=0.95,color=['green','orange'],
label=['men','women'])

plt.legend()
```
